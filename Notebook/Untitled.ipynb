```python
import pandas as pd
```

```python
loc = pd.read_csv('tmcs_locations.csv')
data = pd.read_csv('tmcs_2020_2029.csv')
data[['location_id', 'location']]
location_id = data.location_id.unique()
location = data.location.unique()
loc_obj = {}
for i in range(len(loc)):
    loc_obj[loc.location_id[i]] = loc.location[i]
print(len(loc))
loc_obj
```

```python
num_lanes = {}
for id in location_id:
    num_lanes[id] = loc_obj[id]
    loc_obj.pop(id)
```

```python
num_lanes = pd.DataFrame(num_lanes.items(), columns = ['id', 'location'])
num_lanes2 = pd.DataFrame(loc_obj.items(), columns = ['id', 'location'])
```
