```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.utils import shuffle
```

```python
data = pd.read_csv('tmcs_2020_2029_final.csv')
data = shuffle(data)
data["is_oneway"] = data["is_oneway"].astype(int)
data["is_weekend"] = data["is_weekend"].astype(int)
data["is_holiday"] = data["is_holiday"].astype(int)
```

```python
ind = []
for index, row in data.iterrows():
    if(row.isin([np.nan, np.inf, -np.inf]).any()):
        ind.append(index)
print(ind)
print(len(ind))
```

```python

def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    print(indices_to_keep)
    return df[indices_to_keep].astype(np.float64)


X = data[['location_id', 'year', 'month', 'day', 'time_start_hour',
       'time_start_min', 'time_end_hour', 'time_end_min', 'num_lanes',
       'is_oneway', 'is_weekend', 'is_holiday']]
y = data[['nx', 'sx', 'ex', 'wx']]
X
```

```python
y
```

```python
model = DecisionTreeRegressor()
model.fit(X, y)
```
